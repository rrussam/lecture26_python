```python
# 매직 매서드 연습

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f'{self.name}({self.age}살)'
    def __eq__(self, other):
        if self.name == other.name and self.age == other.age:
            return True
        return False

p1 = Person('홍길동', 15)
print(p1)

p2 = Person('홍길동', 15)

if p1 == p2 :
    print('같은사람')
else :
    print('다른사람')
```

```python
class Integer :
    def __init__(self, value) :
        self.value = value

    def __add__(self, other) :
        return self.value + other.value

    def __sub__(self, other) :
        return self.value - other.value

a = Integer(5)
b = Integer(3)

print(a + b)
print(a - b)
```

```python
class Student :
    def __init__(self, name, school) :
        self.name = name
        self.school = school

    def info(self) :
        print(f'대학 : {self.school} ,  이름 : {self.name}')

student = Student('이성실', '한국대학교')
student.info()
```

```python
class Student :
    GRADE_A = 'A'
    GRADE_B = 'B'
    GRADE_C = 'C'
    GRADE_D = 'D'
    GRADE_F = 'F'
    def __init__(self, name, dept, mid, final) :
        self.name = name
        self.dept = dept
        self.mid = int(mid)
        self.final = int(final)

    def __str__(self) :
        return f'학과 : {self.dept}, 이름 : {self.name}, 중간 : {self.mid}, 기말 : {self.final}'

    def grade(self) :
        average = (self.mid + self.final) / 2
        if average >= 90 :
            return Student.GRADE_A
        elif average >= 80 :
            return Student.GRADE_B
        elif average >= 70 :
            return Student.GRADE_C
        elif average >= 60 :
            return Student.GRADE_D
        else :
            return Student.GRADE_F


student = Student('김경철', '기계학과', '89', '90')
print(student)
print(student.grade())
```
