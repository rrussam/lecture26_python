```python
# 매직 매서드 연습

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f'{self.name}({self.age}살)'
    def __eq__(self, other):
        if self.name == other.name and self.age == other.age:
            return True
        return False

p1 = Person('홍길동', 15)
print(p1)

p2 = Person('홍길동', 15)

if p1 == p2 :
    print('같은사람')
else :
    print('다른사람')
```
