```python
# while문 연습 : 숫자를 입력받아서 총합과 평균을 출력

sum = cnt = 0
num = int(input('숫자 입력 (0을 입력하면 끝) :'))

while num != 0:
    sum += num
    cnt += 1
    num = int(input('숫자 입력 (0을 입력하면 끝) :'))

if cnt != 0 : 
    average = float(sum / cnt)
else :
    average = 0

print(f'>> 총합은 {sum}이고, 평균은 {average : .2f}이다.')
```

```python
# 무한 루프와 분기문 break 활용

sum = cnt = 0

while True:
    num = int(input('숫자 입력 (0을 입력하면 끝) :'))
    if num == 0 : break
    sum += num
    cnt += 1

if cnt != 0 : 
    average = float(sum / cnt)
else :
    average = 0

print(f'>> 총합은 {sum}이고, 평균은 {average : .2f}이다.')
```

```python
# for문 연습 : 숫자를 입력받고 그 숫자에 해당하는 구구단 출력

num = int(input('몇 단을 출력할까? (1~9까지 숫자 입력)'))

if 1 <= num <= 9 :
    for i in range(1,10):
        print(f'{num} x {i} = {num * i}')

else :
    print('범위 밖이라서 구구단을 출력할 수 없어!')
```

```python
# for문 연습 : 숫자를 입력받고 그 수의 팩토리얼을 출력

fac = 1
num = int(input('몇 팩토리얼을 출력할까? :'))

for i in range(2, num + 1) :
    fac *= i

print(f'{num}! = {fac}')
```
