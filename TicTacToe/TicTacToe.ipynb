```python
### data structure
player = 'O'
com = 'X'
board = [x for x in range(1, 10)]
turn = 0
win_status = [(0,1,2), (3,4,5), (6,7,8),
              (0,4,8), (2,4,8),
              (0,3,6), (1,4,7),(2,5,8)]
              

# function definition
def isWin(turn) :
    for status in win_status :
        if board[status[0]] == board[status[1]] == board[status[2]] == turn:
            return True

    return False
    
def ShowBoard() :
    print()
    for i in range(9) :
        print(f'|{board[i]}', end = '')
        if (i + 1) % 3 == 0 :
            print('|')
            print()
    print()

def TypeNum() :
    num = int(input('Type Num :'))
    return num

def GetComNum() :
    print('Computer Turn :')
    for i in range(9) :
        if board[i] == i + 1 :
            return i + 1
    

def MarkBoard(turn, num) :
    board[num - 1] = turn
    

# main logic
print('========= Tic Tac Toe =========')
ShowBoard()
while True :
    
    my_num = TypeNum()
    MarkBoard(player, my_num)
    ShowBoard()
    
    if isWin(player) : 
        print('Player Win!')
        break

    com_num = GetComNum()

    if com_num == -1 :
        break
    else :
        MarkBoard(com, com_num)
    ShowBoard()
    
        
    if isWin(com) : 
        print('Com Win!')
        break
```
