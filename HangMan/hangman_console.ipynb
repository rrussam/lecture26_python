```python
import random
class Hangman :
    MAX_TRY = 7
    MASK_CHAR = '_'
    RIGHT = 1
    WRONG = 0
    EXIST = -1
    WIN = 1
    LOSE = 0
    CONTINUE = -1
    def __init__(self, word_list) :
        self.word = random.choice(word_list).upper()
        self.display_word = Hangman.MASK_CHAR * len(self.word)
        self.num_try = 0
        
    def check_letter(self, letter) :
        letter = letter.upper()
        if self.word.count(letter) > 0 : # 있는 문자면 display_word에서 위치를 찾아 수정
            for i in range(len(self.word)) :
                if self.word[i] == letter :
                    self.display_word = self.display_word[:i] + letter + self.display_word[i+1:]
                
            return Hangman.RIGHT
        else :
            self.num_try += 1
            return Hangman.WRONG
        
    def is_win(self) :
        if self.display_word.count(Hangman.MASK_CHAR) == 0 :
            return Hangman.WIN
        elif self.num_try >= Hangman.MAX_TRY :
            return Hangman.LOSE

        return Hangman.CONTINUE


print()
print('============= Hangman =============')
word_list = ['apple', 'banana', 'man', 'woman', 'tomato']

hangman = Hangman(word_list)
print(f'{hangman.display_word}', f'{len(hangman.word)}글자')
while True :
    letter = input('>> 알파벳 입력 : ')
    result = hangman.check_letter(letter)
    
    if result == Hangman.RIGHT :
        print(f'정답 : {hangman.display_word}')
    elif result == Hangman.WRONG :
        print(f'오답 : {hangman.num_try}회 시도')
    
    
    result = hangman.is_win()
    if result == Hangman.WIN :
        print(f'YOU WIN!!! (TRY : {hangman.num_try})')
        break
    elif result == Hangman.LOSE :
        print(f'YOU LOSE!!! (ANSWER : {hangman.word})')
        break
```
