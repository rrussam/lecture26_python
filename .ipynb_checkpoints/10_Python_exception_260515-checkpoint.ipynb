```python
# try ... except ...

try :
    number = int(input('>> 정수 입력 :'))
    print(number)
except ValueError :
    print('정수를 입력하시오.')
```

```python
num_list = [2, 6, 34, 65, 11]
try :
    index = int(input('>> plz input index :'))
    print(num_list[index])
except (ValueError, IndexError) :
    print(f'plz type integer number which is in 0 to {len(num_list) - 1}')
except Exception as e :
    print(f'Exception Ocurred : {e} -> {type(e)}')
```

```python
try :
    number = int(input('>> plz input number :'))
except ValueError :
    print('no number input')
else :
    print(number)
finally :
    print('end')
```

```python
while True :
    try :
        number = int(input('양의 정수 입력 :'))
        if number <= 0 :
            raise ValueError
    except ValueError :
        print('양의 정수를 입력하세요.')
    else :
        sum = 0
        for i in range(1, number + 1) :
            sum += i
        print(f'>> 1부터 {number}까지의 값 : {sum}')
        break
```

### 학습 내용 되돌아보기

```python
# 1
try :
    f = float('30f')
except Exception as e :
    print(type(e)) # Value Error

# 2
try :
    {'fb' : 11, 'bb' : 9, 'vb' : 6}['foot']
except Exception as e :
    print(type(e)) # Key Error

# 3
try :
    [1, 2, 3][3]
except Exception as e :
    print(type(e)) # Index Error

# 4
try :
    pl = 'python' + 3
except Exception as e :
    print(type(e)) # Type Error
```

```python
try :
    m = int(input('>> input month : '))
    if not (1 <= m <= 12) :
        raise Exception('1 ~ 12 사이의 정수를 입력하세요.')

except Exception as e :
    print('예외 발생 : ', e)

else :
    print(f'{m}월 : 잘 입력했습니다.')
```

```python
def divide(x, y) :
    try :
        if y == 0 :
            raise Exception('0으로는 나눌 수 없습니다')
        result = x / y
    except Exception as e :
        print(f'Exception : {e}')
    else :
        print(f'결과 {result}')

divide(3.2, 2)
divide(5.4, 0)
```

```python
uyear = {1 : 'freshman', 2 : 'sophomore', 3 : 'junior', 4 : 'senior'}

try :
    user_input = int(input('대학교 몇 학년이지요?'))
    if not (1 <= user_input <= 4 ) :
        raise Exception
except Exception as e :
    print(f'예외 발생 이름 : {type(e)}')
    print('예외 발생 이유 : 1 ~ 4 정수를 입력하세요.')
else :
    print(f'{user_input}학년 : {uyear[user_input]}')
finally :
    print('예외 처리가 잘 되는군요!')
```
