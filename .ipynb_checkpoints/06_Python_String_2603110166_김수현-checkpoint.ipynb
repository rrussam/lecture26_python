### f String

```python
string = 'ABC'
print(f"|{string:10}|")
print(f"|{string:>10}|")
print(f"|{string:^10}|")
print(f"|{string:*>10}|")
print(f"|{string:-^10}|")
```

```python
number = 1234
print(f"|{number:20}|")
print(f"|{number:<20}|")
print(f"|{number:-^20}|")
print(f"|{number:*>20}|")
```

### 모스부호 반환코드

```python
str = input('알파벳 문자열 입력 :')
str = str.upper()

mo = ['·-', '-···', '-·-·', '-··', '·', '··-·', '--·', '····', '··', '·---', '-·-', '·-··', '--', '-·', '---', '·--·', '--·-', '·-·', '···', '-', '··-', '···-', '·--', '-··-', '-·--', '--··']

print('모스부호로 변환 :', end = ' ')
for i in range(len(str)) :
    print(mo[ord(str[i]) - 65], end = ' ')
```
