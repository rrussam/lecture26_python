```python
s = input().strip()
part = set()

start = 0
end = 1
l = len(s)

while True :
    part.add(s[start:end])
    end += 1
    if end > l :
        start += 1
        end = start + 1
    if start > l :
        break

print(len(part) - 1)
```

```python
n = int(input())
s = set()

for _ in range(n) :
    name, cmd = input().split()
    if cmd == 'enter' :
        s.add(name)
    else :
        s.remove(name)

result = sorted(s, reverse = True)
for person in result :
    print(person)
```
