### 리스트의 항목에 접근하기 : 슬라이싱(slicing)

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]

print(test_list[0:5])
print(test_list[:5])

print(test_list[1::2])
print(test_list[:5:2])
print(test_list[::3])
print(test_list[:])
print(test_list[::])

print(test_list[-1:-7:-2])
print(test_list[1:-1:])
print(test_list[-3::-1])
print(test_list[::-1])
```

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8]
test_list[:2] = [111, 222, 333]
print(test_list)
```

### 리스트의 역순 및 정렬 저장 : reverse(), sort(), sorted()

```python
a_list = ['one', 'two', 'three', 'four']
print('원본 데이터', a_list)
a_list.reverse()
print('역순 데이터', a_list)

a_list.sort()
print('정렬 데이터', a_list)

a_list = ['one', 'two', 'three', 'four']
b_list = sorted(a_list)
print('원본 데이터', a_list)
print('정렬 데이터', b_list)

c_list = sorted(a_list, key = len)
print(c_list)
```

```python
scores = [95, 45, 21, 74, 55]
print(sorted(scores, reverse = True))

print(scores)
scores.sort(reverse = True)
print(scores)
```

### 리스트의 데이터 추가 : append()

```python
N = int(input())
num_list = []
for i in range(1, N+1) :
    num_list.append(i)
print(num_list)
```

```python
N = int(input())
num_list = [i for i in range(1, N+1)]
print(num_list)

num_list = [i * 2 for i in range(1, N+1)]
print(num_list)

N = int(input())
num_list = [i for i in range(1, N+1) if i % 2 == 0]
print(num_list)
```

### 리스트 항목 수정 : insert(), extend()

```python
list1 = [1,2]
list2 = [5,6,7]

list1.append(4)
list1.insert(2,3)
print(list1)

list1.extend(list2)
print(list1)
```
