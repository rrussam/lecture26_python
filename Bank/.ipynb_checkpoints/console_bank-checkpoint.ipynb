```python
class Account :
    def __init__(self, account_no, owner, balance) :
        self.__account_no = account_no
        self.__owner = owner
        self.__balance = balance
        
    def __str__(self) :
        return f'{self.__account_no}\t{self.__owner}\t{self.__balance}'
        
    def deposit(self, amount) :
        self.__balance += amount
    def withdraw(self, amount) :
        self.__balance -= amount
        
    def get_account_no(self) :
        return self.__account_no
    def get_owner(self) :
        return self.__owner
    def get_balance(self) :
        return self.__balance

class AccountService :
    def __init__(self) :
        self.__account_list = []
        
    def create_account(self, account_number, owner, balance) :
        account = Account(account_number, owner, balance)
        self.__account_list.append(account)

    def list_account(self) :
        return self.__account_list

    def deposit(self, account_number, amount) :
        for account in account_list :
            if account.get_account_no() == account_number :
                account.deposit(amount)
                break

    def withdraw(self, account_number, amount) :
        for account in account_list :
            if account.get_account_no() == account_number :
                account.withdraw(amount)
                break

# 메뉴오 사용자 상호작용에 따른 서비스 호출

def select_menu() :
    print('=======================================================')
    print('1. 계좌 생성 | 2. 계좌 목록 | 3. 입금 | 4. 출금 | 0. 종료')
    print('=======================================================')
    menu = int(input('>> select menu : '))
    return menu

aservice = AccountService()
print()
print(' ================= Susu Bank ================= ')
while True :
    menu = select_menu()
    if menu == 0 :
        break
    elif menu == 1 :
        # 계좌 번호와 계좌주, 잔액을 입력받아서 계좌를 생성
        account_number = input('>> 계좌 번호 : ')
        owner = input('>> 계좌주 : ')
        balance = int(input('>> 초기 입금액 : '))
        if aservice.create_account(account_number, owner, balance) :
            print('계좌가 생성되었습니다.')
    elif menu == 2 : # 계좌 목록
        account_list = aservice.list_account()
        
        print('====================')
        print('===== 계좌 목록 =====')
        print('====================')
        for account in account_list :
            print(account)

    elif menu == 3 :
        print('============')
        print('=== 예금 ===')
        print('============')
        account_number = input('>> 계좌 번호 : ')
        amount = int(input('>> 예금액 : '))
        aservice.deposit(account_number, amount)

    elif menu == 4 :
        print('============')
        print('=== 출금 ===')
        print('============')
        account_number = input('>> 계좌 번호 : ')
        amount = int(input('>> 출금액 : '))
        aservice.withdraw(account_number, amount)

print(' =================   thanks   ================= ')
```
