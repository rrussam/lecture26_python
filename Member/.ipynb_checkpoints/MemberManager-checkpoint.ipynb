```python
class Member :
    LINE_DIV = '-' * 20
    def __init__(self, u_no, u_id, u_pw, u_name, phone, address) :
        self.__u_no = u_no
        self.__u_id = u_id
        self.__u_pw = u_pw
        self.__u_name = u_name
        self.__phone = phone
        self.__address = address

    def __str__(self) :
        return f'{Member.LINE_DIV}\nMember Number : {self.__u_no} \nID : {self.__u_id} \nPassword : {self.__u_pw} \nUser Name : {self.__u_name} \nPhone Number : {self.__phone} \nAddress : {self.__address} \n{Member.LINE_DIV}'

    def UserUpdate(self, u_name, u_pw, phone, address) :
        self.__u_name = u_name
        self.__u_pw = u_pw
        self.__phone = phone
        self.__address = address

    def GetUserNumber(self) :
        return self.__u_no

class MemberService :
    def __init__(self) :
        self.__user_list = []

    # 사용자를 멤버에 추가 시도
    def AddUser(self, u_no, u_id, u_pw, u_name, phone, address) :
        try :
            nUser = Member(u_no, u_id, u_pw, u_name, phone, address)
            if nUser in self.__user_list :
                raise ExistingError

        except ExistingError :
            # 유저 정보가 이미 존재함
            return False

        else :
            self.__user_list.append(nUser)
            return True

    # 사용자 목록 반환
    def UserList(self) :
        return self.__user_list

    # 사용자 상세 정보 표시
    def UserInfo(self, u_id) :
        for user in self.__user_list :
            if user.GetUserNumber() == u_id :
                print(user)
                return True
        return False

    # 사용자 정보 업데이트
    def UserUpdate(self, u_no, u_id, u_pw) :
        pass


    # 사용자 탈퇴
    def DeleteUser(self, u_no, u_id, u_pw) :
        pass


    # 사용자 존재 여부
    def user_exist(self) :
        if len(self.__user_list) > 0 :
            return True
        return False


def SelectMenu() :
    print('==========================================================================================')
    print(' 1. register | 2. user list | 3. user detail | 4. edit user | 5. Delete Account | 0. Exit')
    print('==========================================================================================')
    select = int(input(''))
    return select
```

```python
mService = MemberService()

while True :
    try :
        select = SelectMenu()

    except ValueError :
        print('plz input integer number.')

    else :
        if select == 0 :
            break

        if select == 1 :
            u_no = int(input('user number :'))
            u_id = input('user id :')
            u_pw = input('user password :')
            u_name = input('user name :')
            phone = input('phone number :')
            address = input('address :')
            if mService.AddUser(u_no, u_id, u_pw, u_name, phone, address) :
                print('User Add : Success')
            else :
                print('Error - something wrong')

        if select == 2 :
            if mService.user_exist() :
                user_list = mService.UserList()
                for user in user_list :
                    print(user)
            else :
                print('no user exist')

        if select == 3 :
            if mService.user_exist() :
                try :
                    u_id = int(input('uid :'))
                    
                except ValueError :
                    print('Value Error')

                else :
                    if mService.UserInfo(u_id) :
                        print('Info show')
                    else :
                        print('no user info')

            else :
                print('no user exist')
```
