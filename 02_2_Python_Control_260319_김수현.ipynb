```python
# for문 연습 : 숫자를 입력받고 그 숫자에 해당하는 구구단 출력

num = int(input('몇 단을 출력할까? (1~9까지 숫자 입력)'))

if 1 <= num <= 9 :
    for i in range(1,10):
        print(f'{num} x {i} = {num * i}')

else :
    print('범위 밖이라서 구구단을 출력할 수 없어!')
```

```python
# 별찍기 0
num = int(input())
for i in range(num) :
    for j in range(num) :
        print('*', end = '')
    print()
```

```python
# 별찍기
num = int(input())
for ln in range(1, num + 1) :
    for star in range(ln) :
        print('*', end = '')
    print('')
```

```python
# 별 거꾸로 찍기
num = int(input())
for ln in range(1, num + 1) :
    for star in range(num - ln + 1) :
        print('*', end = '')
    print('')
```

```python
# 별 앞에 공백

num = int(input())
for ln in range(num) :
    for emp in range(num - (ln + 1)) :
        print(' ', end = '')
    for star in range(ln + 1) :
        print('*', end = '')
    print('')
```

```python
# 별 앞에 공백 2

num = int(input())
for ln in range(num) :
    for emp in range(ln) :
        print(' ', end = '')
    for star in range(num - ln) :
        print('*', end = '')
    print('')
```

```python
num = int(input())
level = num * 2

for ln in range(1, level, 2) :
    
    space = int((level - ln) / 2)
    
    for emp in range(space) :
        print(' ', end = '')
        
    for star in range(ln) :
        print('*', end = '')
    
    print('')
```

```python
num = int(input())
level = num * 2

for ln in range(1, level, 2) :
    
    space = int(ln / 2)
    
    for emp in range(space) :
        print(' ', end = '')
        
    for star in range(level - ln) :
        print('*', end = '')
        
    print('')
```
