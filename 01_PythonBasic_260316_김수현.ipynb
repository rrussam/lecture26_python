### 02. 대입연산자
* 두 변수값 서로 바꾸기 (swap)

```python
a = int(input("첫번째 수 입력 : "))
b = int(input("두번째 수 입력 : "))
print(a, b, " 입력한 두 수")
swap = a
a = b
b = swap
print(a, b, " 서로 바뀐 두 수")
```

```python
a = b = c = 7
print(a, b, c)
```

```python
a, b, c = 2, 5, 8
print(a, b, c)
```

```python
a = 5
b = 8
print(a, b)
a, b = b, a
print(a, b)
```

```python
a = 10 > 20
print(a)
```

```python
# 원의 넓이 구하기

#반지름 입력
r = int(input('반지름 입력 : '))

#원주율 입력
pi = 3.14159265358979

#입력받은 값으로 넓이 계산 및 출력
area = float(r * r * pi)
print(f'반지름이 {r}인 원의 넓이는 {area:.3f}입니다.')
```
